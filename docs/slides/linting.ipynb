## What is linting?

Linting is the automated checking of your source code for programmatic and stylistic errors. This is done by a linting tool. A linting tool is static code analyzer. 

The goal of linting is to ensure code quality, maintainability, and consistency.

## Key Aspects of Linting

1. **Error Detection:** Linting tools identify potential errors in the code, such as syntax mistakes or unused variables.
2. **Coding Standards:** Linting enforces coding style guidelines, such as consistent indentation, naming conventions, or line lengths, to make the code more readable and standardized.
3. **Bug Prevention:** By highlighting problematic patterns, linting helps prevent bugs from making it into production.
4. **Static Analysis:** Linting is a form of static code analysis, meaning it checks the code without executing it.

## Linting workflow

1. **Source Code Input:** The linter takes the source code as input.
2. **Analysis:** The tool parses the code and compares it against a set of rules or patterns defined in its configuration.
3. **Feedback:** It outputs a list of warnings, errors, or suggestions for improving the code.

## Styleguide

- PEP 8: <https://peps.python.org/pep-0008/>
- Google Python Styleguide: <https://google.github.io/styleguide/pyguide.html>
- Mozilla Python Styleguide: <https://firefox-source-docs.mozilla.org/code-quality/coding-style/coding_style_python.html>
- Using your own style (ruff rules): <https://docs.astral.sh/ruff/rules/>

```toml
# <https://github.com/Netflix/iceberg-python/blob/main/ruff.toml>
[tool.ruff.lint]
select = [
    "E", # pycodestyle
    "W", # pycodestyle
    "F", # Pyflakes
    "B", # flake8-bugbear
    "PIE", # flake8-pie
    "C4", # flake8-comprehensions
    "I", # isort
    "UP", # pyupgrade
]
ignore = ["E501","E203","B024","B028","UP037"]

# Allow autofix for all enabled rules (when `--fix`) is provided.
fixable = ["ALL"]
unfixable = []

per-file-ignores = {}

# Allow unused variables when underscore-prefixed.
dummy-variable-rgx = "^(_+|(_+[a-zA-Z0-9_]*[a-zA-Z0-9]+?))$"
```
