# OpenTelemetry

- <https://opentelemetry.io/>
- Goal: High-quality, ubiquitous, and portable telemetry to enable effective observability
- Collection of SDKs, APIs and tools to instrument, generate, collect and export telemetry data (metrics, logs and traces) to analyze software’s performance and behavior

![alt text](./images/otel.PNG)

## Monitoring

- Monitoring is the process of collecting, analyzing, and visualizing system metrics and logs to track the health and performance of applications and infrastructure.
- "How is the system performing?"
- **reactive**

## Observability
- Observability is the capability to understand the internal state of a system based on the data it produces, enabling faster root cause analysis and informed decision-making.
- "Why is the system performing that way?"
- **pro-active**


## Telemetry: Metrics, Logs and Traces

- Metrics: A measurement captured at runtime.
  - Error rate, CPU usage, memory usage, average response time, etc.
- Logs: A recording of a event.
  - Warning, error, info, debug, etc.
- Traces: Path of request through your Application.
  - API calls, database queries, internal logic, etc.

## Reliability: SLI, SLO, SLA

- Reliability: Is the system doing what it is supposed to do? 
- Service Level Indicators (SLI): Is the system doing it well?
- Service Level Objectives (SLO): Is the system doing it well enough?
- Service Level Agreements (SLA): Is the system doing it well enough for the customer?

## Instrumentation

- The process of modifying your application to collect telemetry data.
- OpenTelemetry provides two ways to instrument your application:
  - Zero-code instrumentation via agents or agent-like installations
    - Monitoring the edge of your application
    - Supported languages: .NET, Go, Java, JavaScript, PHP, Python
  - Code-based instrumentation via OpenTelemetry SDKs and APIs
    - Monitoring the inside of your application
    - Supported languages: C++, .NET, Erlang/Elixir, Go, Java, JavaScript, PHP, Python, Ruby, Rust, Swift, ...

## Four golden signals

- Latency: The time it takes to service a request.
- Traffic: A measure of how much demand is being placed on your system, measured in a high-level system-specific metric.
- Errors: The rate of requests that fail.
- Saturation: How "full" your service is.
